```python
segm1 = []
segm2 = []

for i in range(0, len(seg1), 2):
    temp = tuple([int(seg1[i]), int(seg1[i+1])])
    segm1.append(temp)

for i in range(0, len(seg2), 2):
    temp = tuple([int(seg2[i]), int(seg2[i+1])])
    segm2.append(temp)
```

```python
width = 640
height = 480

img = Image.new('L', (width, height), 0)
ImageDraw.Draw(img).polygon(segm1, outline=1, fill=1)
mask = np.array(img)

img2 = Image.new('L', (width, height), 0)
ImageDraw.Draw(img2).polygon(segm2, outline=1, fill=1)
mask2 = np.array(img2)
```

```python
for i in range(0,480):
    for j in range(0,640):
        if mask[i][j] == 1:
            ImageDraw.Draw(img).point(xy=[i, j], fill=255)
        if mask2[i][j] == 1:
            ImageDraw.Draw(img).point(xy=[i, j], fill=150)
img.show()
```

```python
intersection = np.logical_and(mask, mask2)
union = np.logical_or(mask, mask2)
iou_score = np.sum(intersection) / np.sum(union)

print("IoU = ", iou_score)
```
